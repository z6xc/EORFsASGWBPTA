```python
import sympy as sp
from sympy import symbols, exp, integrate, oo, E1, Ei,Ci,Si, expint, log, gamma, latex
from sympy import EulerGamma
from sympy import nsimplify
import os
```

```python
def mkdir(path):
    import os
    path=path.strip()
    path=path.rstrip("\\")
    isExists=os.path.exists(path)
    if not isExists:
        os.makedirs(path) 
        print (path+' 创建成功')
        return path
    else:
        print (path+' 目录已存在')
        return path
```

```python
Rlm_Def_All_SM_eq_smp_Savepath = mkdir(os.getcwd()+"/Rlm_IJ_Duan_SM_eq_smp")
```

```python
def Rlm_IJ_gamma_duan(Lty,duan,args,LM):
    avt,aut,acmt,gammat = args
    acm, au, av, gamma = sp.symbols('a_{-} a_u a_v gamma')
    Rlm_Def_All_SM_nam_smp = "/Rlm_{}_{}.txt".format(duan,str(Lty)+"_"+"(a3, a2, acm, gamma)"+"_"+str(LM))
    Savepath = Rlm_Def_All_SM_eq_smp_Savepath
    Rlm_path = Savepath+Rlm_Def_All_SM_nam_smp#load_formula_as_function(Savepath+Rlm_Def_All_SM_nam_smp)
    with open(Rlm_path, 'r', encoding='utf-8') as f:
                Rlm_str = f.read().strip()
    Rlm_str = Rlm_str.replace("EulerGamma","euler_gamma")#.replace("au",str(aut))
    #Rlm_str = Rlm_str#.replace("acm",str(acmt)).replace("gamma",str(gammat))
    #Rlm = Rlm_str
    Rlm = eval(Rlm_str)
    Rlm = nsimplify(Rlm)
    Rlm = Rlm.subs({av:avt,au:aut,acm:acmt,gamma:gammat})
    return Rlm#(au,av,gamma)
```

```python
def Rlm_IJ_gamma_duan_num(Lty,duan,args,LM):
    avnum,aunum,gammanum = args
    acmnum = sp.sqrt(-2*aunum*avnum*sp.cos(gammanum) + aunum**2 + avnum**2).evalf()
    argsnum = avnum,aunum,acmnum,gammanum
    Rlm = Rlm_IJ_gamma_duan(Lty,duan,argsnum,LM)
    #Rlm = Rlm.subs({sp.pi:x/sp.pi}).subs({x:sp.pi})/sp.pi
    Rlm = Rlm.subs({euler_gamma:EulerGamma})
    Rlm_num = Rlm.evalf()
    return Rlm_num
```

```python
a3, a2, gamma = sp.symbols('a3 a2 gamma')
au, av, gamma = sp.symbols('a_u a_v gamma')
acm = sp.symbols('a_{-}')
euler_gamma = sp.symbols('gamma_E')
```
